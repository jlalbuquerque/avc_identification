```python
import numpy as np
import cv2

# Carregando as imagens PNG processadas
img_ct = cv2.imread('/home/gpds/Documentos/CT-500Info/PACIENTES COM AVCI VERSÃO 3/train/1/CT 219_CT000000.png', cv2.IMREAD_GRAYSCALE)
img_base2 = cv2.imread('/home/gpds/Documentos/CT-500Info/PACIENTES COM AVCI VERSÃO 3/train/1/PACIENTE 223Paciente223 (15).png', cv2.IMREAD_GRAYSCALE)

# Calcular estatísticas para a base CT
ct_mean = np.mean(img_ct)
ct_median = np.median(img_ct)
ct_std = np.std(img_ct)
ct_min = np.min(img_ct)
ct_max = np.max(img_ct)

# Calcular estatísticas para a base 2
base2_mean = np.mean(img_base2)
base2_median = np.median(img_base2)
base2_std = np.std(img_base2)
base2_min = np.min(img_base2)
base2_max = np.max(img_base2)

# Exibir os resultados
print("CT Statistics:")
print(f"Mean: {ct_mean}, Median: {ct_median}, Std Dev: {ct_std}, Min: {ct_min}, Max: {ct_max}")

print("\nBase 2 Statistics:")
print(f"Mean: {base2_mean}, Median: {base2_median}, Std Dev: {base2_std}, Min: {base2_min}, Max: {base2_max}")
```

```python
import matplotlib.pyplot as plt

# Criar histogramas
plt.figure(figsize=(12, 6))

# Histograma da imagem CT
plt.subplot(1, 2, 1)
plt.hist(img_ct.ravel(), bins=256, color='gray', alpha=0.7)
plt.title('Histograma da Imagem CT')
plt.xlim([0, 255])

# Histograma da base 2
plt.subplot(1, 2, 2)
plt.hist(img_base2.ravel(), bins=256, color='gray', alpha=0.7)
plt.title('Histograma da Base 2')
plt.xlim([0, 255])

plt.show()
```

```python
offset = 20
img_ct_adjusted = img_ct - offset
# Garantir que os valores não se tornem negativos e que não excedam 255
img_ct_adjusted = np.clip(img_ct_adjusted, 0, 255).astype(np.uint8)
img_ct_adjusted = img_ct_adjusted*255/np.max(img_ct_adjusted)


# Salvar a imagem ajustada
cv2.imwrite('adjusted_ct_image.png', img_ct_adjusted)

# Visualizar a imagem ajustada
cv2.imshow('Imagem CT Ajustada', img_ct_adjusted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Visualizar as imagens originais e ajustadas
plt.figure(figsize=(12, 6))

# Imagem CT original
plt.subplot(1, 3, 1)
plt.imshow(img_ct, cmap='gray')
plt.title('Imagem CT Original')
plt.axis('off')

# Imagem Base 2
plt.subplot(1, 3, 2)
plt.imshow(img_base2, cmap='gray')
plt.title('Imagem Base 2')
plt.axis('off')

# Imagem CT Ajustada
plt.subplot(1, 3, 3)
plt.imshow(img_ct_adjusted, cmap='gray')
plt.title('Imagem CT Ajustada')
plt.axis('off')

plt.show()
```
